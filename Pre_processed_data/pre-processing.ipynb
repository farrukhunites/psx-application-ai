```python
import pandas as pd
output_filename = "sys_data.csv"
df = pd.read_csv(f'../Stocks/{output_filename}')
df
```

```python
num_zero_in_high_price = (df['High Price'] == 0.00).sum()
num_zero_in_low_price = (df['Low Price'] == 0.00).sum()
num_zero_in_closing_price = (df['Closing Price'] == 0.00).sum()

print(f'Number of rows with 0 in High Price: {num_zero_in_high_price}')
print(f'Number of rows with 0 in Low Price: {num_zero_in_low_price}')
print(f'Number of rows with 0 in Closing Price: {num_zero_in_closing_price}')
rows, columns = df.shape
print(f'Number of rows: {rows}')
print(f'Number of columns: {columns}')

```

Filtered DataFrame for Type 'REG'

```python
# Step 1: Remove extra single quotes from the Type column
df['Type'] = df['Type'].str.replace("'", "", regex=False)

# Step 2: Now filter for rows where Type is 'REG'
df_reg = df[df['Type'] == 'REG']

# Step 3: Print the filtered DataFrame
print("Filtered DataFrame for Type 'REG':")
print(df_reg)

num_zero_in_high_price = (df_reg['High Price'] == 0.00).sum()
num_zero_in_low_price = (df_reg['Low Price'] == 0.00).sum()
num_zero_in_closing_price = (df_reg['Closing Price'] == 0.00).sum()

# Step 4: Print the counts of zero values
print(f'Number of rows with 0 in High Price for Type "REG": {num_zero_in_high_price}')
print(f'Number of rows with 0 in Low Price for Type "REG": {num_zero_in_low_price}')
print(f'Number of rows with 0 in Closing Price for Type "REG": {num_zero_in_closing_price}')

# Step 5: Count the number of rows with Type 'REG'
num_rows_with_reg = df_reg.shape[0]

# Print the number of rows with Type 'REG'
print(f'Number of rows with Type "REG": {num_rows_with_reg}')

df_reg.to_csv(output_filename, index=False)

```

GRAPHS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot (High Price vs Low Price)
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_reg, x='High Price', y='Low Price', hue='Closing Price', palette='viridis', alpha=0.7)
plt.title('Scatter Plot of High Price vs Low Price (Type REG)')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.legend(title='Closing Price', loc='upper right')
plt.show()

# Scatter Plot (High Price vs Closing Price)
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_reg, x='High Price', y='Closing Price', hue='Low Price', palette='coolwarm', alpha=0.7)
plt.title('Scatter Plot of High Price vs Closing Price (Type REG)')
plt.xlabel('High Price')
plt.ylabel('Closing Price')
plt.legend(title='Low Price', loc='upper right')
plt.show()

```
