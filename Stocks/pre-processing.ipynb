CORRECTION  OF DATE FORMAT

```python
# Step 1: Remove extra single quotes from the Type column
df['Type'] = df['Type'].str.replace("'", "", regex=False)

# Step 2: Now filter for rows where Type is 'REG'
df_reg = df[df['Type'] == 'REG'].copy()  # Make a copy of the filtered DataFrame

# Step 3: Clean the Date column
# Assuming the Date column is named 'Date'
df_reg['Date'] = df_reg['Date'].str.extract(r'(\d{4}-\w{3}-\d{2})')[0]

# Step 4: Convert cleaned date strings to datetime format using .loc
df_reg.loc[:, 'Date'] = pd.to_datetime(df_reg['Date'], format='%Y-%b-%d', errors='coerce')

# Step 5: Print the filtered DataFrame
print("Filtered DataFrame for Type 'REG':")
print(df_reg)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (Assuming df is already defined)
# Step 1: Remove extra single quotes from the Type column
df['Type'] = df['Type'].str.replace("'", "", regex=False)

# Step 2: Now filter for rows where Type is 'REG'
df_reg = df[df['Type'] == 'REG'].copy()  # Make a copy of the filtered DataFrame

# Step 3: Clean the Date column
df_reg['Date'] = df_reg['Date'].str.extract(r'(\d{4}-\w{3}-\d{2})')[0]

# Step 4: Convert cleaned date strings to datetime format using .loc
df_reg.loc[:, 'Date'] = pd.to_datetime(df_reg['Date'], format='%Y-%b-%d', errors='coerce')

# Ensure 'Closing Price' column is numeric
df_reg['Closing Price'] = pd.to_numeric(df_reg['Closing Price'], errors='coerce')

# Step 5: Remove rows with NaT or NaN
df_reg.dropna(subset=['Date', 'Closing Price'], inplace=True)

# Step 6: Plotting the line plot
plt.figure(figsize=(12, 6))
plt.plot(df_reg['Date'], df_reg['Closing Price'], marker='o', linestyle='-', color='b', label='Closing Price')
plt.title('Closing Price Over Time for Type REG')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
